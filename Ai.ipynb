```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("plant.jpeg", cv2.IMREAD_COLOR)
plt.imshow(img)
```

```python
img2 = cv2.imread("plant.jpeg", cv2.IMREAD_GRAYSCALE)
```

```python
sobel = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
plt.imshow(img2)
```

```python
plt.subplot(1,1,1),plt.imshow(sobel,cmap = 'gray')
```
